## A notebook to create a pivot table for the HRApop cover

## Install and import libraries

## Load data

```python
hra_pop_version = "v0.11.1"

df = pd.read_csv(f"../../../hra-pop/output-data/{hra_pop_version}/reports/atlas/validation-v6.csv")
df
```

## Build sheet for pivot

```python
# filter by canonical cell types via ASCT+B tables, then select only CTs in male, left kidney
# filtered_df = df[(df['asct_relation_in_asctb_table'] | df['indirect_asct_relation_in_asctb_table']) & (df['sex'] == 'Male' | df['tool'] == 'azimuth')]

df_filtered = df[(df['asct_relation_in_asctb_table'] |
                  df['indirect_asct_relation_in_asctb_table']) & (df['sex'] == 'Male') & (df['tool'] == 'azimuth') & (df['organ'] == 'left kidney')]

df_filtered
```

## Export CSV

```python
df_filtered.to_csv("output/pivot.csv", index=False)
```
