## Set Up Environment

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re
import umap
%matplotlib inline
```

```python
# Figure settings
sns.set_theme(context='notebook', 
              style='white', 
              rc={'figure.figsize':(14,10)},
              font='sans-serif', font_scale=1, color_codes=True)
```

## Loading data

```python
data = pd.read_csv("https://raw.githubusercontent.com/cns-iu/hra-cell-type-populations-supporting-information/main/ad_hoc_queries/output/umap_as_percentage.csv")
data.head()
```

## Preparing Categorical Variables

```python
# Creates variable for biological sex.
data['sex'] = np.where(data['organ'].str.contains('VHF'),'female','male')
data.sex = data.sex.astype("category")
```

```python
# Removes VHM & VHF from organ names
data.organ = data.organ.str.replace("VHF","", regex=True)
data.organ = data.organ.str.replace("VHM","", regex=True)
data.organ = data.organ.str.replace("Left","", regex=True)
data.organ = data.organ.str.replace("Right","", regex=True)
data.organ = data.organ.astype("category")
data['organCode'] = data['organ'].cat.codes
```

```python
# Creates object with list of organs
data_organs = data[['dataset_combinations','organ','organCode']].drop_duplicates()
```

```python
# Creates dataset for UMAP
data_umap = data.iloc[0:282, 3:43]
```

## UMAP Analysis

```python
scaled_data_umap = StandardScaler().fit_transform(data_umap)
```

```python
# Version 1 - Euclidean Distance
reducer = umap.UMAP(n_neighbors=10, min_dist=1, n_components=2, metric='euclidean')
embedding = reducer.fit_transform(scaled_data_umap)
embedding.shape
```

```python
sns.scatterplot(x = embedding[:, 0], 
                y = embedding[:, 1], s=25,
                hue=data_organs.organ, alpha=0.7)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), title="Organ", labels=['Colon','Heart','Kidney','Large Intestine','Liver',
                   'Lung','Mammary Gland','Prostate','Skin','Small Intestine','Spleen','Uterer','Urinary Bladder'])
plt.title('UMAP Results for HRA-Pop Data Samples Tissue Percentages', loc='left', pad=8, fontsize=20)
```

```python
# Version 2 - Manhattan Distance
reducer = umap.UMAP(n_neighbors=25, min_dist=1, n_components=2, metric='manhattan')
embedding2 = reducer.fit_transform(scaled_data_umap)
embedding2.shape
```

```python
# UMAP Scatter Plot of Results
sns.scatterplot(x = embedding2[:, 0], 
                y = embedding2[:, 1], s=25,
                hue=data_organs.organ, alpha=0.7)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), title="Organ", labels=['Colon','Heart','Kidney','Large Intestine','Liver',
                   'Lung','Mammary Gland','Prostate','Skin','Small Intestine','Spleen','Uterer','Urinary Bladder'])
plt.title('UMAP Results for HRA-Pop Data Samples Tissue Percentages', loc='left', pad=8, fontsize=20)
```
